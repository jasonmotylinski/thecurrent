```python
import csv
from rapidfuzz import process, fuzz
choices = []
with open('output/thecurrent/csv/2025/12/20251231.csv', 'r') as f:
    reader = csv.DictReader(f)
    for row in reader:
        choices.append(row['artist'].strip())

with open('output/kutx/csv/2025/12/20251231.csv', 'r') as f:
    reader = csv.DictReader(f)
    for row in reader:
        match=process.extractOne( row['artist'].strip(), choices, scorer=fuzz.WRatio)
        if(match[1]>=91 and match[1]<100):
            print(f"{row['artist']} => {match[0]} ({match[1]})")
```
