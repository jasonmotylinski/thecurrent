```python
import requests
import bs4 as bs
from datetime import datetime, timedelta, timezone

resp=requests.get('https://wfuv.org/views/ajax?created[min]=2024-01-02&created[max]=2024-01-02&view_name=on_air_playlist&view_display_id=block_wfuv_on_air_playlist')


tbody=bs.BeautifulSoup(resp.json()[3]['data']).select_one('tbody')
_date=datetime.strptime('2024-01-01', "%Y-%m-%d")
for tr in tbody.select('tr'):
    dte=tr.find('td', {'headers': 'view-created-table-column'}).text.strip()
    song=tr.find('td', {'headers': 'view-title-table-column'}).text.strip()
    artist=tr.find('td', {'headers': 'view-field-artist-table-column'}).text.strip()
    d=datetime.strptime(dte, "%m/%d, %I:%M%p")
    played_at=d.replace(year=int(_date.strftime("%Y")), tzinfo = timezone(offset=timedelta(hours=-5)))
    print({"song": song, "artist": artist, "played_at": played_at})
```
