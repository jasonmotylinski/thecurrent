```python
import pandas as pd
import requests
import sqlite3
import wikipedia

from bs4 import BeautifulSoup
```

```python
t='SELECT artist, COUNT(*) as ct FROM songs GROUP BY artist ORDER BY ct DESC LIMIT 100'

WIKIPEDIA_URL="https://en.wikipedia.org/wiki/{topic}"
WIKIPEDIA_BAND_URL="https://en.wikipedia.org/wiki/{topic}_(band)"
WIKIPEDIA_MUSICIAN_URL="https://en.wikipedia.org/wiki/{topic}_(musician)"
WIKIPEDIA_MUSIC_GROUP_URL="https://en.wikipedia.org/wiki/{topic}_(music_group)"

URLS ={
    "bio": WIKIPEDIA_URL, 
    "band": WIKIPEDIA_BAND_URL,
    "musician": WIKIPEDIA_MUSICIAN_URL,
    "music group": WIKIPEDIA_MUSIC_GROUP_URL
}


con=sqlite3.connect("thecurrent.sqlite3")
df = pd.read_sql(t, con)

def get_vcard(url, artist):
    r=requests.get(url.format(topic=artist))
    soup=BeautifulSoup(r.text)
    bio=soup("table", {"class": "vcard"})
    if len(bio) > 0:
        return bio
    return None

for r in df.iterrows():
    artist=r[1]["artist"]

    for k in URLS.keys():
        vcard=get_vcard(URLS[k], artist)
        if vcard:
            print("{0} has {1} vard".format(artist, k))
            break
    
    if not vcard:
        print("{0} does not have a vcard".format(artist))

```
